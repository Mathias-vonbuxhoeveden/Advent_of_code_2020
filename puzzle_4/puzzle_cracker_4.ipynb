```python
#Read in data

lines = open('input_data.csv').read().splitlines()
```

```python
lines.append('')
```

```python
parsed_data = []
new_string = ""

for x in lines:   
     
    new_string = new_string + " " + x
    
    if x == "":
        
        parsed_data.append(new_string)
        new_string = ""  
```

```python
def collect_fields(input_string):
    
    input_string = input_string.strip()
    input_string = input_string.split(' ')
    keys = []
    
    
    for string in input_string:
        
        keys.append(string.split(':')[0])
        
        
    return keys
```

```python
def collect_data(input_string):
    
    input_string = input_string.strip()
    input_string = input_string.split(' ')
    keys = []
    
    
    for string in input_string:
        
        keys.append(string.split(':')[1])
        
        
    return keys
```

```python
def check_if_input_is_valid(collected_fields):
    
    mandatory_fields = ['byr','iyr','eyr','hgt','hcl','ecl','pid']
    check = all(elem in collected_fields  for elem in mandatory_fields)
    
    if check:
        
        return 1
    
    else:
        
        return 0
```

```python
def check_if_data_is_valid(collected_fields, collected_data):
    
    def validate_birth_year(birth_year):

        mask1 = int(birth_year) >= 1920
        mask2 = int(birth_year) <= 2002
        mask3 = (mask1 & mask2)

        return mask3
    
    def validate_issue_year(issue_year):
        
        mask1 = int(issue_year) >= 2010
        mask2 = int(issue_year) <= 2020
        mask3 = (mask1 & mask2)
        
        return mask3
    
    def validate_expiration_year(experation_year):
        
        mask1 = int(experation_year) >= 2020
        mask2 = int(experation_year) <= 2030
        mask3 = (mask1 & mask2)
        
        return mask3
    
    def validate_experation_year(experation_year):
        
        mask1 = int(experation_year) >= 2020
        mask2 = int(experation_year) <= 2030
        mask3 = (mask1 & mask2)
        
        return mask3
    
    def validate_height(heigth):
        print(heigth)
        measure = heigth[-2:]
        
        if measure == 'cm':
            
            mask1 = int(heigth[:-2]) >= 150
            mask2 = int(heigth[:-2]) <= 193
            mask3 = (mask1 & mask2)
            
            return mask3
            
        elif measure == 'in':
            
            mask1 = int(heigth[:-2]) >= 59
            mask2 = int(heigth[:-2]) <= 76
            mask3 = (mask1 & mask2)
            
            return mask3
        
        else:
            
            return False
        
    def validate_hair_color(hair_color):
        
        mask1 = hair_color[0] == '#'
        hair_color[1:] 
        
        occurence_count = np.sum([hair_color.count('0'),
                                  hair_color.count('1'),
                                  hair_color.count('2'),
                                  hair_color.count('3'),
                                  hair_color.count('4'),
                                  hair_color.count('5'),
                                  hair_color.count('6'),
                                  hair_color.count('7'),
                                  hair_color.count('8'),
                                  hair_color.count('9'),
                                  hair_color.count('9'),
                                  hair_color.count('a'),
                                  hair_color.count('b'),
                                  hair_color.count('c'),
                                  hair_color.count('d'),
                                  hair_color.count('e'),
                                  hair_color.count('f')])
        
        mask2 = occurence_count == 6
        mask3 = (mask1 & mask2)
        
        return mask3
    
    def validate_eye_color(eye_color):
        
        return eye_color in ['amb', 'blu', 'brn', 'gry', 'grn', 'hzl', 'oth']
    
    def validate_passport_number(passport_number):
        
        mask1 = passport_number.isnumeric()
        mask2 = len(passport_number) == 9
        mask3 = (mask1 & mask2)
        
        return mask3        
        
    
    
    data_dict = dict(zip(collected_fields, collected_data))
    birth_year_valid = validate_birth_year(birth_year = data_dict['byr'])
    issue_year_valid = validate_issue_year(issue_year = data_dict['iyr'])
    experation_year_valid = validate_expiration_year(experation_year = data_dict['eyr'])
    height_valid = validate_height(heigth = data_dict['hgt'])
    hair_color_valid = validate_hair_color(hair_color = data_dict['hcl'])
    eye_color_valid = validate_eye_color(eye_color = data_dict['ecl'])
    passport_number_valid = validate_passport_number(passport_number = data_dict['pid'])
    
    mask = (birth_year_valid & issue_year_valid & experation_year_valid & height_valid & hair_color_valid & eye_color_valid & passport_number_valid)
    
    if mask:
        
        return 1
    
    else:
        
        return 0
    
```

```python
result = []

for input_data in parsed_data:

    collected_fields = collect_fields(input_string = input_data)
    result.append(check_if_input_is_valid(collected_fields = collected_fields))    
```

```python
print("The data contains {} valid passports.".format(np.sum(result)))
```
