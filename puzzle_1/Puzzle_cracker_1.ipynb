```python
def find_target_sum(input_data, target_sum):
    
    for first_value in input_data:
        
        for second_value in input_data:
            
            for third_value in input_data:
                
                test_sum = first_value + second_value + third_value
                
                if test_sum == target_sum:
                
                    return (first_value, second_value, third_value)
            
                else:
                
                    pass
```

```python
result = find_target_sum(input_data = values, target_sum = 2020)
```

```python
product = result[0]*result[1]*result[2]
```

```python
print("The result is {}".format(product))
```
