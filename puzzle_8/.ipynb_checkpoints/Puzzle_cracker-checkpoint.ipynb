```python
import pandas as pd
```

```python
def load_data():
    
    df = pd.read_csv('input_data.csv', header = None)
    df.columns = ['data']
    df = df.data.tolist()
    
    return df    
```

```python
def get_action_at_positon(input_data, accumulator):
    
    action_type = input_data[0:3]
    action_number = input_data[4:]
    postion_incrementer = 1
    
    if action_number[0] == "+":
        
        action_number = int(action_number[1:])
        
    else:
        
        action_number = -1*int(action_number[1:])
        
    
    if action_type == "acc":
        
        accumulator += action_number
        
    elif action_type == "jmp":
        
        postion_incrementer += action_number
        
    else:
        
        pass    
    
    return (accumulator, postion_incrementer) 
```
